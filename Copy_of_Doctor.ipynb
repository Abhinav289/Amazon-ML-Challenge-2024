```python
import torch
from doctr.io import DocumentFile
from doctr.models import ocr_predictor
```

```python
import os
```

```python
def load_predictor(det_arch: str = 'db_resnet50', reco_arch: str = 'crnn_vgg16_bn'):
    """Load the OCR predictor from doctr.models"""
    predictor = ocr_predictor(det_arch, reco_arch, pretrained=True)
    return predictor
```

```python
def process_image(image_path: str) -> list:
    """Process an image and return a list of recognized words"""
    # Load the image
   # image = DocumentFile.from_images('pragyaan rover.png')
    image = DocumentFile.from_images(image_path)

    # Load the OCR model
    predictor = load_predictor()

    # Perform OCR on the image
    result = predictor(image)

    # Extract words
    words = []
    for page in result.pages:
        for block in page.blocks:
            for line in block.lines:
                for word in line.words:
                    words.append(word.value)

    return words

```

```python
import re

def convert_to_centimetre(text):
   # print(f"Converting to centimetre: {text}")
    text = text.lower()
    text = re.sub(r'(\d+)\s*cm', r'\1 centimetre', text)
    text = re.sub(r'(\d+)\s*mm', r'\1 milimetre', text)
    text = re.sub(r'(\d+)\s*m', r'\1 metre', text)
    text = re.sub(r'(\d+)\s*ft', r'\1 foot', text)
    text = re.sub(r'(\d+)\s*in', r'\1 inch', text)
    text = re.sub(r'(\d+)\s*yd', r'\1 yard', text)

    range_match = re.match(r'\[(\d+\.?\d*),\s*(\d+\.?\d*)\]\s*(centimetre|foot|millimetre|metre|inch|yard)', text, re.IGNORECASE)
    if range_match:
        min_value = float(range_match.group(1))
        unit = range_match.group(3).lower()
        conversion = {'centimetre': 1, 'foot': 30.48, 'millimetre': 0.1, 'metre': 100, 'inch': 2.54, 'yard': 91.44}
        min_cm = min_value * conversion.get(unit, 1)
        return '{} centimetre'.format(min_cm)

    single_value_match = re.match(r'(\d+\.?\d*)\s*(centimetre|foot|millimetre|metre|inch|yard)', text, re.IGNORECASE)
    if single_value_match:
        value = float(single_value_match.group(1))
        unit = single_value_match.group(2).lower()
        conversion = {'centimetre': 1, 'foot': 30.48, 'millimetre': 0.1, 'metre': 100, 'inch': 2.54, 'yard': 91.44}
        value_cm = value * conversion.get(unit, 1)
        return '{} centimetre'.format(value_cm)

    large_value_match = re.match(r'(\d+\.?\d*)\s*(e\d+)\s*(millimetre|centimetre|metre)', text, re.IGNORECASE)
    if large_value_match:
        value = float(large_value_match.group(1)) * 10 ** float(large_value_match.group(2))
        unit = large_value_match.group(3).lower()
        conversion = {'centimetre': 1, 'millimetre': 0.1, 'metre': 100}
        value_cm = value * conversion.get(unit, 1)
        return '{} centimetre'.format(value_cm)

    return ''

def convert_to_volt(text):
   # print(f"Converting to volt: {text}")
    text = text.lower()
    text = re.sub(r'(\d+)\s*v', r'\1 volt', text)
    text = re.sub(r'(\d+)\s*mv', r'\1 milivolt', text)
    text = re.sub(r'(\d+)\s*kv', r'\1 kilovolt', text)

    range_match = re.match(r'\[(\d+\.?\d*),\s*(\d+\.?\d*)\]\s*(volt|millivolt|kilovolt)', text, re.IGNORECASE)
    if range_match:
        max_value = float(range_match.group(2))
        unit = range_match.group(3).lower()
        conversion = {'volt': 1, 'millivolt': 0.001, 'kilovolt': 1000}
        max_volt = max_value * conversion.get(unit, 1)
        return '{} volt'.format(max_volt)

    single_value_match = re.match(r'(\d+\.?\d*)\s*(volt|millivolt|kilovolt)', text, re.IGNORECASE)
    if single_value_match:
        value = float(single_value_match.group(1))
        unit = single_value_match.group(2).lower()
        conversion = {'volt': 1, 'millivolt': 0.001, 'kilovolt': 1000}
        value_volt = value * conversion.get(unit, 1)
        return '{} volt'.format(value_volt)

    large_value_match = re.match(r'(\d+\.?\d*)\s*(e\d+)\s*(volt|millivolt|kilovolt)', text, re.IGNORECASE)
    if large_value_match:
        value = float(large_value_match.group(1)) * 10 ** float(large_value_match.group(2))
        unit = large_value_match.group(3).lower()
        conversion = {'volt': 1, 'millivolt': 0.001, 'kilovolt': 1000}
        value_volt = value * conversion.get(unit, 1)
        return '{} volt'.format(value_volt)

    return ''

def convert_hp_to_watts(text):
   # print(f"Converting horsepower to watts: {text}")
    text = text.lower()
    text = re.sub(r'(\d+)\s*hp', r'\1 horsepower', text, flags=re.IGNORECASE)
    text = re.sub(r'(\d+)\s*kw', r'\1 kilowatt', text, flags=re.IGNORECASE)
    text = re.sub(r'(\d+)\s*w', r'\1 watt', text, flags=re.IGNORECASE)
   # print(text)
    single_value_match = re.match(r'(\d+\.?\d*)\s*(horsepower|kilowatt|watt)', text, re.IGNORECASE)
   # print(single_value_match)
    if single_value_match:
        value = float(single_value_match.group(1))
        unit = single_value_match.group(2).lower()
        conversion = {'horsepower': 745.7, 'kilowatt': 1000, 'watt': 1}
        value_watts = value * conversion.get(unit, 1)
        return '{} watts'.format(value_watts)

    range_match = re.match(r'\[(\d+\.?\d*),\s*(\d+\.?\d*)\]\s*(horsepower|kilowatt|watt)', text, re.IGNORECASE)
    if range_match:
        max_value = float(range_match.group(2))
        unit = range_match.group(3).lower()
        conversion = {'horsepower': 745.7, 'kilowatt': 1000, 'watt': 1}
        max_watts = max_value * conversion.get(unit, 1)
        return '{} watts'.format(max_watts)

    large_value_match = re.match(r'(\d+\.?\d*)\s*(e\d+)\s*(horsepower|kilowatt|watt)', text, re.IGNORECASE)
    if large_value_match:
        value = float(large_value_match.group(1)) * 10 ** float(large_value_match.group(2))
        unit = large_value_match.group(3).lower()
        conversion = {'horsepower': 745.7, 'kilowatt': 1000, 'watt': 1}
        value_watts = value * conversion.get(unit, 1)
        return '{} watts'.format(value_watts)



def convert_to_grams(text):
    # print(f"Converting to grams: {text}")
    text = text.lower()
    text = re.sub(r'(\d+)\s*kg', r'\1 kilogram', text)
    text = re.sub(r'(\d+)\s*lb', r'\1 pound', text)
    text = re.sub(r'(\d+)\s*oz', r'\1 ounce', text)
    text = re.sub(r'(\d+)\s*mg', r'\1 milligram', text)
    text = re.sub(r'(\d+)\s*mcg', r'\1 microgram', text)
    text = re.sub(r'(\d+)\s*ng',r'\1 nanogram', text)
    text = re.sub(r'(\d+)\s*tonne', r'\1 ton', text)

    range_match = re.match(r'\[(\d+\.?\d*),\s*(\d+\.?\d*)\]\s*(kilogram|pound|ounce|gram|ton)', text, re.IGNORECASE)
    if range_match:
        min_value = float(range_match.group(1))
        unit = range_match.group(3).lower()
        conversion = {'kilogram': 1000, 'pound': 453.592, 'ounce': 28.3495, 'ton': 1e6, 'gram': 1}
        min_grams = min_value * conversion.get(unit, 1)
        return '{:.2f} gram'.format(min_grams)

    range_match = re.match(r'(\d+\.?\d*)\s*(kilogram|pound|ounce|gram|ton)\s*to\s*(\d+\.?\d*)\s*(kilogram|pound|ounce|gram|ton)', text, re.IGNORECASE)
    if range_match:
        min_value = float(range_match.group(1))
        unit = range_match.group(2).lower()
        min_grams = min_value * conversion.get(unit, 1)
        return '{:.2f} gram'.format(min_grams)

    large_value_match = re.match(r'(\d+\.?\d*)\s*(e\d+)\s*(milligram|microgram|nanogram)', text, re.IGNORECASE)
    if large_value_match:
        value = float(large_value_match.group(1)) * 10 ** float(large_value_match.group(2))
        unit = large_value_match.group(3).lower()
        conversion = {'milligram': 0.001, 'microgram': 1e-6, 'nanogram': 1e-9}
        value_grams = value * conversion.get(unit, 1)
        return '{:.2f} gram'.format(value_grams)

    single_unit_match = re.match(r'(\d+\.?\d*)\s*(kilogram|pound|ounce|gram|ton|milligram|microgram|nanogram)', text, re.IGNORECASE)
    if single_unit_match:
        value = float(single_unit_match.group(1))
        unit = single_unit_match.group(2).lower()
        conversion = {'kilogram': 1000, 'pound': 453.592, 'ounce': 28.3495, 'ton': 1e6, 'milligram': 0.001, 'microgram': 1e-6, 'nanogram': 1e-9, 'gram': 1}
        value_grams = value * conversion.get(unit, 1)
        return '{:.2f} gram'.format(value_grams)

    return ''

def master_conversion_function(input_list, entity_type):
   # print(f"Input List: {input_list}, Entity Type: {entity_type}")

    conversion_functions = {
        'width': convert_to_centimetre,
        'depth': convert_to_centimetre,
        'height': convert_to_centimetre,
        'item_weight': convert_to_grams,
        'maximum_weight_recommendation': convert_to_grams,
        'voltage': convert_to_volt,
        'wattage': convert_hp_to_watts,
        'item_volume': convert_to_grams,  # Adjust this based on actual volume conversion needs
    }

    conversion_function = conversion_functions.get(entity_type)
    if not conversion_function:
        # print(f"No conversion function for entity type: {entity_type}")
        return ''


    for item in input_list:
       # print(f"Processing item: {item}")
        result = conversion_function(item)
       # print(f"Conversion result: {result}")
        if result:
            #print(f"Conversion result: {result}")
            return result




# # Example usage
# input_list = ['24W', '32ng', '21m', 'lorem ipsum']
# entity_type="wattage"
# result=master_conversion_function(input_list, entity_type)  # Should output '24 watt'
```
